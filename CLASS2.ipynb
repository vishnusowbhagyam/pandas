```python
### merge(): help to merge the dataframes, based upon common column
df1 = pd.DataFrame({'ID' : [1,2,3], 'Hero' : ['Mahesh Babu','Omesh','Manoj']})
df2 = pd.DataFrame({'ID' : [2,3,4], 'MovieFee' : [250, 10, 50]})

merged_df = pd.merge(df1, df2, on = 'ID')
merged_df
```

```python
#inner: it returns only common columns from both tables
merged_df = pd.merge(df1, df2, on = 'ID', how = 'inner')
merged_df
```

```python
#outer: return all data from left and right but common data return only once
merged_df = pd.merge(df1, df2, on = 'ID', how = 'outer')
merged_df
```

```python
#right: return all data from right and common from left
merged_df = pd.merge(df1, df2, on = 'ID', how = 'right')
merged_df
```

```python
#left: all the data from left table and common from right
merged_df = pd.merge(df1, df2, on = 'ID', how = 'left')
merged_df
```

### 10. Time Series operations in pandas

```python
date_rng = pd.date_range(start = "2020-01-01",end = '2025-12-31',freq = 'h')
date_rng
```

```python
import numpy as np
np.random.seed(42)
random_dates = np.random.choice(date_rng, size = 1000, replace = False)
random_dates
```

```python
# creating dataframe based upon above time series data
df = pd.DataFrame({
    'timestamp' : pd.to_datetime(random_dates), # convert numpy datetime to pandas datetime format
    'order_id' : np.arange(1, 1001), # generating numbers 1 to 1000
    'user_id' : np.random.randint(1000, 5001, size = 1000), # generate 1000 integer b/w 1000 - 5001
    'category': np.random.choice(['Biryani','Haleem','Maggie','Crab Masala','Chole Bhature'], size = 1000),
    'price' : np.random.randint(100,500, size = 1000),
    'quantity' : np.random.randint(1,5, size = 1000)
})
df
```

```python
df['sales'] = df['price'] * df['quantity']
```

```python
df = df.sort_values('timestamp').reset_index(drop = True)
df
```

```python
df['Year'] = df.timestamp.dt.year
df['Month'] = df.timestamp.dt.month
df['Day'] = df.timestamp.dt.day
df['Hour'] = df.timestamp.dt.hour
df['Minute'] = df.timestamp.dt.minute
df['Second'] = df.timestamp.dt.second
df
```
